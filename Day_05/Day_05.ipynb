```python
#Level 2:

ages = [19, 22, 19,24,20, 25, 26, 24, 25, 24]
ages.sort()
print(ages)

min_ages = min(ages) 
max_ages = max(ages)
ages.append(min_ages)
ages.append(max_ages)

print(ages)

print(len(ages))

median_item = (ages[5] + ages[6])/2
print(median_item)

average  = (sum(ages))/len(ages)
print(average)

ranges_of_ages = max(ages) - min(ages)
print(ranges_of_ages)

comparison = (abs(min_ages)- average) > (abs(max_ages)-average)
print(comparison)
```
